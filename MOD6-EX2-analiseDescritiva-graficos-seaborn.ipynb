```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import do CSV
df = pd.read_csv("sinasc_RO_2019.csv")

# gero outro Data Frame só com as colunas que vou usar
df = df[ ['IDADEMAE', 'PESO'] ].copy()

print(df.shape)

df.head()
```

```python
df.describe().transpose()
```

```python
df.isnull().sum()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.close('all')

ax = sns.boxplot(data = df, y = 'IDADEMAE')

ax.set_title("Idade das mães")
ax.set_ylabel("")

plt.show()

#OBS: NOTA-SE ALGUNS OUTLIERS NO LIMITE SUPERIOR
```

####  2) Faça um boxplot do peso do bebê

```python
plt.close('all')

ax = sns.boxplot(data = df, y = 'PESO')

ax.set_title("Peso dos bebês")
ax.set_ylabel("")

plt.show()

# MUITOS OUTLIERS, NOS DOIS LIMITES
```

####  3) Faça um histograma da idade da mãe

```python
plt.close('all')

ax = sns.histplot(data = df, x = 'IDADEMAE', bins=20, kde=True)

ax.set_xlabel("Idade das mães")
ax.set_ylabel("Quantidade")


plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.close('all')

ax = sns.histplot(data = df, x = 'PESO', bins=20, kde=True)

ax.set_xlabel("Peso dos bebês")
ax.set_ylabel("Quantidade")


plt.show()
```
