````python
df_metadados = pd.DataFrame()

df_metadados['variavel'] = list(df.columns)

df_metadados['tipo_dado'] = list(df.dtypes)

df_metadados
````

````python
tot_linhas, tot_colunas = df.shape


lista_qtde_categorias = []


for i in range(0, tot_colunas):
    
    qtde = len(df.iloc[:,i].unique())
    lista_qtde_categorias.append(qtde)
    
    
df_metadados['qtd_categorias'] = lista_qtde_categorias

df_metadados
````

#### 3) Crie variáveis dummy para as variáveis necessárias (i.e. aquelas que são qualitativas e não estão armazenadas como {0, 1} ou {True, False}. Crie um *dataframe* apenas com as variáveis apropriadas para entrada no scikitlearn - elimine as variáveis tipo *str*, mantendo apenas suas versões *dummy*.

````python
df['qt_pessoas_residencia'].value_counts()
````

````python
df_tratado = pd.DataFrame()

# booleans para int
df_tratado['mau'] = df['mau'].astype(int)

# numericas discretas de dois dominios
df_tratado['possui_fone_comercial'] = df['possui_fone_comercial']
df_tratado['possui_fone'] = df['possui_fone']
df_tratado['possui_email'] = df['possui_email']

# categoricas de duas dominios
df_tratado['sexo'] = df['sexo'].map( {'M': 0, 'F': 1} )
df_tratado['posse_de_veiculo'] = df['posse_de_veiculo'].map( {'N': 0, 'Y': 1} )
df_tratado['posse_de_imovel'] = df['posse_de_imovel'].map( {'N': 0, 'Y': 1} )

# numericas discretas com mais de dois valores
df_tratado['qtd_filhos'] = df['qtd_filhos']
df_tratado['qt_pessoas_residencia'] = df['qt_pessoas_residencia'].astype(int)

# categóricas com mais de dois dominios
df_tratado['tipo_renda'] = df['tipo_renda']
df_tratado['educacao'] = df['educacao']
df_tratado['estado_civil'] = df['estado_civil']
df_tratado['tipo_residencia'] = df['tipo_residencia']


# numericas continuas com muitos valores
df_tratado['idade'] = df['idade']
df_tratado['tempo_emprego'] = df['tempo_emprego']



df_tratado = pd.get_dummies(data = df_tratado, columns=['qtd_filhos', 'qt_pessoas_residencia', 'tipo_renda', 
                                                        'educacao', 'estado_civil', 'tipo_residencia'])


df_tratado.head()
````

#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
# tabela cruzada / freq do "possui_email"

tab_cruzada = pd.crosstab(index = df['possui_email'], columns=df['mau'])
tab_cruzada.div( tab_cruzada.sum(1), axis=0 )
````

````python
# tabela cruzada / freq do "posse_de_veiculo"

tab_cruzada = pd.crosstab(index=df['posse_de_veiculo'], columns=df['mau'])
tab_cruzada.div(tab_cruzada.sum(1), axis=0)
````

Nenhuma das variaveis "possui_email" e "posse_de_veiculo" é mais poderosa que a outra pra prever o caso mau=1. 

Como visto nas tabelas cruzadas, os casos são proporcionais.
