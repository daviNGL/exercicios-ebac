Neste exercício vamos trabalhar com a base "Heart Disease Data Set" disponível no site do [UCI](https://archive.ics.uci.edu/ml/datasets/heart+disease).

Temos 303 indivíduos e vamos utilizar 13 variáveis disponíveis para prever a presença de doença cardíaca na variável ```num```(a 14$^a$ variável).

A descrição das variáveis está recortada abaixo:
- age: idade do paciente em anos
- sex: sexo (1 = male; 0 = female)  
- cp: tipo de dor no peito
  - 1: angina típica
  - 2: angina atípica
  - 3: dor não-angina
  - 4: assintomático
- trestbps: pressão sanguínea em repouso (em mm Hg na admissão ao hospital
- chol: colesterol sérico em mg/dl
- fbs: (açúcar no sangue em jejum > 120 mg/dl) (1 = True; 0 = False)
- restecg: resultados eletrocardiográficos em repouso
  - 0: normal
  - 1: tendo anormalidade da onda ST-T (Inversões de onda T e / ou ST com elevação ou depressão de > 0.05 mV)
  - 2: mostrando hipertrofia ventricular esquerda provável ou definitiva pelos critérios de Estes
- thalach: frequência cardíaca máxima alcançada
- exang: angina induzida por exercício(1 = sim; 0 = não)
- oldpeak = Depressão de ST induzida por exercício em relação ao repouso
- slope: Depressão de ST induzida por exercício em relação ao repouso
  - 1: inclinação ascendente
  - 2: estável
  - 3: inclinação descendente
- ca: número de vasos principais (0-3) coloridos por fluorosopia
- thal: 3 = normal; 6 = defeito corrigido; 7 = defeito reversível
- num: diagnóstico de doença cardíaga (status de doença angiográfica)

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import statsmodels.formula.api as smf
````

````python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data'

df = pd.read_csv(url, 
                 names=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg',
                        'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num'])
df.head()
````

````python
df['flag_doente'] = (df['num'] != 0).astype('int64')
df
````

1. Monte uma bivariada para a variável ```sex``` com as seguintes estatísticas:

- Número de *saudáveis*
- Número de *doentes*
- Total
- Quantidade média de doentes
- *Odds*
- *Odds ratio* (Use o total como referência)
- Logito (LOG(*Odds*))
- O LOG do *Odds ratio* (com o total como referência) é chamado de *Weight of Evidence* (WOE). Calcule-o.

**Criação da tabela cruzada**

````python
tab_cruzada_sex = pd.crosstab(index = df['sex'], columns = df['flag_doente'], margins=True)

# Ajusto o nome das linhas e colunas

tab_cruzada_sex.index = ['feminino', 'masculino', 'total']

tab_cruzada_sex.columns = ['nao_doente', 'doente', 'total']

tab_cruzada_sex
````

**Cálculo das métricas**

````python
# Probabilidade de determinado sexo estar doente: qtde_doentes(no sexo) / qtd_total
tab_cruzada_sex['probabilidade'] = tab_cruzada_sex['doente'] / tab_cruzada_sex['total']

# Canche (odds): porcentagem evento / porcentagem não evento
tab_cruzada_sex['chance'] = tab_cruzada_sex['doente'] / tab_cruzada_sex['nao_doente']

# Razão de chances (odds ratio): razão entre as chances
tab_cruzada_sex['razao_chances'] = tab_cruzada_sex['chance'] / tab_cruzada_sex.loc['total', 'chance']

# Logito: log(odds)
tab_cruzada_sex['logito'] = np.log(tab_cruzada_sex['chance'])

# WOE: log(odds ratio)
tab_cruzada_sex['woe'] = np.log(tab_cruzada_sex['razao_chances'])

tab_cruzada_sex
````

2. Monte uma bivariada para a variável ```age``` com as seguintes estatísticas:
- Categorize a variável em 5 grupos
- Calcule as mesmas quantidades calculadas no execício anterior
- Calcule também o valor médio da variável ```age```

**Categorização da variável age**

````python
df['age_cat'] = pd.qcut(x = df['age'], q = 5)

df.head()
````

**Criação da tabela cruzada**

````python
tab_cruzada_age = pd.crosstab(index = df['age_cat'],  columns = df['flag_doente'], margins = True)

# Ajusto o nome das linhas e colunas
tab_cruzada_age.columns = ['nao_doente', 'doente', 'total']

tab_cruzada_age.index = ['29-45', '45-53', '53-58', '58-62', '62-77', 'total']

tab_cruzada_age
````

**Cálculo das métricas**

````python
# Probabilidade de determinada idade estar doente: qtde_doentes(no sexo) / qtd_total
tab_cruzada_age['probabilidade'] = tab_cruzada_age['doente'] / tab_cruzada_age['total']

# Canche (odds): porcentagem evento / porcentagem não evento
tab_cruzada_age['chance'] = tab_cruzada_age['doente'] / tab_cruzada_age['nao_doente']

# Razão de chances (odds ratio): razão entre as chances
tab_cruzada_age['razao_chances'] = tab_cruzada_age['chance'] / tab_cruzada_age.loc['total', 'chance']

# Logito: log(odds)
tab_cruzada_age['logito'] = np.log(tab_cruzada_age['chance'])

# WOE: log(odds ratio)
tab_cruzada_age['woe'] = np.log(tab_cruzada_age['razao_chances'])

tab_cruzada_age
````

3. Visualize a informação da bivariada construída acima através de um gráfico da sua escolha.

````python
figura, (ax1, ax2) = plt.subplots(1, 2, figsize=(15,5))


# Barplot idade X doente
sns.barplot(data = tab_cruzada_age, 
            x = tab_cruzada_age.index.drop('total'), 
            y = tab_cruzada_age['doente'].drop('total'),
            ax = ax1)

ax1.set_title('Quantidade de doentes por faixa de idade')



# Pointplot idade x probabilidade doente
sns.pointplot(x = tab_cruzada_age.index.drop('total'), 
             y = tab_cruzada_age['probabilidade'].drop('total'), 
             ax = ax2)

ax2.set_title('Probabilidade de estar doente por faixa de idade')


plt.show()
````
